```python
import yt_dlp
import pathlib

# --- URL del episodio de YouTube ---
YOUTUBE_URL = "https://www.youtube.com/watch?v=LJiUDxj-2ZE"  # url del episodio

# --- Carpeta donde guardar los audios ---
AUDIO_DIR = pathlib.Path("/content/audios")
AUDIO_DIR.mkdir(parents=True, exist_ok=True)

# --- Configuración de descarga ---
ytdl_opts = {
    "format": "bestaudio/best",
    "outtmpl": str(AUDIO_DIR / "%(title)s.%(ext)s"),
    "postprocessors": [{
        "key": "FFmpegExtractAudio",
        "preferredcodec": "mp3",
        "preferredquality": "192",
    }],
}

# --- Descargar ---
with yt_dlp.YoutubeDL(ytdl_opts) as ydl:
    info = ydl.extract_info(YOUTUBE_URL, download=True)
    audio_path = AUDIO_DIR / f"{info['title']}.mp3"

print(f"[OK] Audio descargado en: {audio_path}")
```

```python
# =========================================
# Ejecutar whisper-diarization con video de youtube
# - Transcripción + diarización
# - Guarda .srt y .txt en /content/resultados
# =========================================
import os, sys, shutil, pathlib, subprocess, time

# --- 1) Configura aquí ---
#AUDIO         = "/content/audios/audio01.mp3"
AUDIO = str(audio_path)
REPO_DIR      = "/content/whisper-diarization"      # repo ya clonado
WHISPER_MODEL = "small"
ENTRYPOINT = "diarize.py"
RESULTS_DIR   = "/content/resultados"               # carpeta de salida

# --- 2) Comprobaciones básicas ---
audio_p = pathlib.Path(AUDIO)
if not audio_p.exists():
    raise FileNotFoundError(f"No existe el audio: {audio_p}. "
                            f"Asegúrate de haberlo movido a {audio_p.parent}.")

repo_p = pathlib.Path(REPO_DIR)
if not repo_p.exists():
    raise FileNotFoundError(f"No encuentro el repo en {repo_p}. Vuelve a clonar o ajusta REPO_DIR.")

# Crear carpeta de resultados
pathlib.Path(RESULTS_DIR).mkdir(parents=True, exist_ok=True)

# --- 3) Ejecutar el pipeline ---
os.chdir(REPO_DIR)
print(f"[INFO] Ejecutando en: {pathlib.Path.cwd()}")
cmd = [
    sys.executable, "-u", ENTRYPOINT,
    "--whisper-model", WHISPER_MODEL,
    "-a", str(audio_p)
]
print("[CMD]", " ".join(cmd))

# Entorno con flags útiles de depuración (opcional)
env = os.environ.copy()
env["PYTHONFAULTHANDLER"]   = "1"   # backtrace si crashea en C/CUDA
env["CUDA_LAUNCH_BLOCKING"] = "1"   # sincroniza kernels CUDA (logs más claros)

# Log en tiempo real + guardarlo a fichero
logfile = "/content/run_diarization.log"
with open(logfile, "w", encoding="utf-8") as lf:
    proc = subprocess.Popen(cmd, env=env, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)
    for line in proc.stdout:
        print(line, end="")
        lf.write(line)
    ret = proc.wait()

print(f"\n[INFO] Return code: {ret}")
if ret != 0:
    print(f"[WARN] El proceso terminó con código {ret}. Revisa {logfile}.")
    # Sugerencias rápidas:
    print(" - Prueba con ENTRYPOINT='diarize.py' (secuencial) si estabas en 'diarize_parallel.py'.")
    print(" - Prueba con WHISPER_MODEL='small'.")
    print(" - Si ves error de 'libcudnn_ops_infer.so.8', asegúrate de usar CTranslate2==4.5.0 o instala libcudnn8.\n")
    raise SystemExit(ret)

# --- 4) Recoger salidas .srt y .txt y copiarlas a /content/resultados ---
base = audio_p.with_suffix("").name     # nombre de archivo sin extensión
srt  = pathlib.Path(REPO_DIR) / f"{base}.srt"
txt  = pathlib.Path(REPO_DIR) / f"{base}.txt"

copiados = []
for f in (srt, txt):
    if f.exists():
        dst = pathlib.Path(RESULTS_DIR) / f.name
        shutil.copy2(f, dst)
        copiados.append(str(dst))

print("\n[OK] Proceso completado.")
print("[OK] Salidas copiadas a:", *copiados, sep="\n  - ")

# (Opcional) Empaquetar resultados
zip_out = pathlib.Path(RESULTS_DIR) / f"{base}.zip"
shutil.make_archive(str(zip_out.with_suffix("")), "zip", RESULTS_DIR)
print(f"[OK] ZIP con resultados: {zip_out}")
```
